```python
# import dependencies

import matplotlib.pyplot as plt
import statistics
import numpy as np
```

```python
x = [50, 42, 39, 34, 32, 48, 22, 17, 8, 13]
y = [17, 19, 21, 22, 24, 23, 34, 31, 43, 35]
```

```python
plt.scatter(x, y, color = 'green', marker = 'D', s = x)
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare($)')
plt.title('Average Fare vs Total Rides')
plt.show()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis, color = 'red', label = 'Chicago', s = y_axis)
plt.xlabel('Fares($)')
plt.ylabel('Months')
plt.legend()
plt.grid()
plt.title('PyBer Fare by Month')
plt.gca().invert_yaxis()
plt.xlim(0, 50)

plt.show()
```

```python
plt.scatter(x_axis, y_axis, s = y_axis, c= 'green')
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
    
plt.scatter(x_axis, y_axis, s= y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s= [i *3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis,label = 'Boston', c = 'skyblue', alpha = 0.8, marker = 'o', edgecolors = 'black', s = [i *5 for i in y_axis])
ax.invert_yaxis()
ax.set_xlabel('Fares($)')
ax.set_ylabel('Months')
ax.set_title('PyBer Fare by Month')
ax.set_xlim(0,50)
ax.legend()
plt.show()
```

```python
colors= ['Red', 'Blue', 'Green', 'Purple', 'Gray', 'SkyBlue', 'Yellow', 'Cyan', 'Orange', 'Pink', 'lightblue', 'Violet']
explode = [0, 0, .3, 0, 0, 0, 0.3, 0, 0, 0, 0, 0]
plt.subplots(figsize= (8, 8))
plt.pie(y_axis, labels = x_axis, counterclock = False, startangle = 98, shadow = True, autopct ='%1.1f%%', explode = explode, colors = colors)

plt.show()
```

```python
stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis, y_axis, yerr= stdev, capsize = 3)
```

```python
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
plt.show()
```

```python
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(2.5))
plt.savefig('../Pyber_Analysis/scatterpractice.png')
plt.show()
```
