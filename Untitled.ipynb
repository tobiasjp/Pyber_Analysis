```python
%matplotlib inline
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# create the plot and add a label for the legend
plt.plot(x_axis, y_axis, label = 'Boston', color="red", linewidth=1, marker = '*')

# create labels for the x and y axes
plt.xlabel('Date')

plt.ylabel('Fare($)')

# Set the y limit between 0 and 45
plt.ylim(0, 50)

# Create a Title
plt.title('Pyber Fare by Month')

#add a legend
plt.legend()

# add a grid
plt.grid()
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, label = 'Boston', color = 'green', marker = 'D', linewidth = 2,)
ax.legend()
ax.set_xlabel('Dates')
ax.set_ylabel('Fares($)')
ax.set_ylim(0, 45)
ax.grid()
ax.set_title('Pyber Fare by Month')
```

```python
# create the plot
plt.bar(x_axis, y_axis)
```

```python
plt.bar(x_axis, y_axis, color = 'green', label = 'Boston')

# create labels for the x and y axes
plt.xlabel('Months')

plt.ylabel('Fares($)')

# create a title
plt.title('PyBer Fares by Month')

# Add a legend
plt.legend()
```

```python
plt.barh(y_axis, x_axis, label="Boston", color ='Pink')
plt.xlabel('Months')
plt.ylabel('Fares($)')
plt.legend()
```

```python
plt.barh(x_axis, y_axis, label = 'Boston', color= 'Yellow')
plt.xlabel('Months')
plt.ylabel('Fares($)')
plt.gca().invert_yaxis()
```

```python
plt.barh(x_axis, y_axis, color = 'magenta', label = 'Boston')
plt.xlabel('Months')
plt.ylabel('Fares($)')
plt.legend()
plt.grid()
plt.gca().invert_yaxis()
```

```python
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color = 'cyan', label = 'Detroit')
ax.set_xlabel('Months')
ax.set_ylabel('Fares($)')
ax.set_title('PyBer Fares by Time')
ax.legend()
ax.grid()
ax.invert_yaxis()
```
